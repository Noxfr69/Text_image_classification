# Edamam Food and Grocery Database
**There are three APIs**
- The first one is to get general information about food items
- the second one is to get nutrients information from a foodID
- the third one is for auto suggestion

We will use the first API

```python
import requests

url = "https://edamam-food-and-grocery-database.p.rapidapi.com/api/food-database/v2/parser"

querystring = {"nutrition-type":"cooking","category[0]":"generic-foods","health[0]":"alcohol-free"}

headers = {
	"X-RapidAPI-Key": "817925db89msh8026458f456141bp1069c1jsn84c757fae491",
	"X-RapidAPI-Host": "edamam-food-and-grocery-database.p.rapidapi.com"
}

response = requests.get(url, headers=headers, params=querystring)

print(response.json())
```

**Here is how the data is structured:**
- text:
- parsed:
- hints:
    - food:
    - measures
- _links:

The data we want : `'foodId', 'label', 'category', 'foodContentsLabel', 'image'` are all under food

```python
import pandas as pd

data = response.json()
# Extract first 10 hints
hints = data['hints'][:10]

# Initialize list to hold data
df_data = []

# Iterate over the hints
for hint in hints:
    # Extract the food data
    food_data = hint['food']
    # Extract the required information
    foodId = food_data.get('foodId', None)
    label = food_data.get('label', None)
    category = food_data.get('category', None)
    foodContentsLabel = food_data.get('foodContentsLabel', None)
    image = food_data.get('image', None)
    # Append to list
    df_data.append([foodId, label, category, foodContentsLabel, image])

# Create DataFrame
df = pd.DataFrame(df_data, columns=['foodId', 'label', 'category', 'foodContentsLabel', 'image'])
```
